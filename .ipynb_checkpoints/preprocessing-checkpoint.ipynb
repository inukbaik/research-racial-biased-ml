```python
import pandas as pd

# Load the dataset
file_path = 'us_racial.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
data.info()
```

```python
# Convert 'Data' to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')
```

```python
# Replace NaN values with 0 across the entire dataset
data = data.fillna(0)
data.head()
```

```python
# Check for the absence of NaN Values
data.isna().sum()
```

```python
# Check for and remove duplicate rows, if any
data = data.drop_duplicates()
```

```python
data.to_csv('data_processed.csv')
```
