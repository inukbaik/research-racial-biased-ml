```python
import pandas as pd

# Load the dataset
file_path = 'deaths.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
data['Location'] = data['County Name'] + ', ' + data['State']
data.tail()
```

```python
data.drop('State', axis=1, inplace=True)
data.drop('County Name', axis=1, inplace=True)
data.head()
```

```python
data.columns
```

```python
new_order = ['Data as of', 'Start Date', 'End Date', 'Location', 'Urban Rural Code',
       'FIPS State', 'FIPS County', 'FIPS Code', 'Indicator', 'Total deaths',
       'COVID-19 Deaths', 'Non-Hispanic White', 'Non-Hispanic Black',
       'Non-Hispanic American Indian or Alaska Native', 'Non-Hispanic Asian',
       'Non-Hispanic Native Hawaiian or Other Pacific Islander', 'Hispanic',
       'Other', 'Urban Rural Description']
data = data[new_order]
data.head()
```

```python
data.to_csv('data_county.csv')
```

```python
file_path2 = 'income_per_capita.csv'
data2 = pd.read_csv(file_path2)
data2.head()
```

```python
data2['2020'] = pd.to_numeric(data2['2020'], errors='coerce')
data2['2021'] = pd.to_numeric(data2['2021'], errors='coerce')
data2['2022'] = pd.to_numeric(data2['2022'], errors='coerce')

data2['Income_per_Capita'] = data2[['2020', '2021', '2022']].mean(axis=1)
data2.head()
```

```python
data2.to_csv('income_per_capita_proc.csv')
```
